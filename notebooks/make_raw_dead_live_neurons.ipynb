Creates the raw images for the dead vs live neuron dataset. The raw brightfield images can be downloaded from:

https://downloads.lbb.ethz.ch/Data/lbb_raw_images.h5

```python
path_raw_images = '../Data/Neuron_Dead_Live/Raw/lbb_raw_neuron_images.h5'
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import h5py as h5

import random

import imageio as io

from skimage.filters import gaussian
```

```python
f = h5.File(path_raw_images,"r")
```

```python
data   = f['Raw/data'][...]/(2**16-1.)
```

```python
np.max(data)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(data[1,:,:,0],cmap='gray')
plt.show()
```

```python
for i in range(data.shape[0]):
    io.imsave('../Data/Neuron_Dead_Live/Raw/'+str(i).zfill(4)+'.png',data[i,:,:,0])
```
